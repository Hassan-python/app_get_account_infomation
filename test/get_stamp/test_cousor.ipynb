```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'MS Gothic'  # 日本語フォントの設定
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# ランダムなデータフレームを生成（5列、100行）
np.random.seed(42)
df = pd.DataFrame({
    'A列': np.random.normal(0, 1, 100),
    'B列': np.random.normal(0, 1, 100),
    'C列': np.random.normal(0, 1, 100),
    'D列': np.random.normal(0, 1, 100),
    'E列': np.random.normal(0, 1, 100)
})

# 相関係数行列を計算
corr_matrix = df.corr()

# ヒートマップを作成
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('相関係数ヒートマップ')
plt.show()
```

```python
# 二次関数 y = ax^2 + bx + c のグラフを描画する

# 二次関数の係数を設定
a = 1
b = 0
c = 0

# xの範囲を設定
x = np.linspace(-10, 10, 400)

# 二次関数の計算
y = a * x**2 + b * x + c

# グラフを描画
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f'y = {a}x² + {b}x + {c}')
plt.title('二次関数のグラフ')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.grid(True)
plt.legend()
plt.show()

```
